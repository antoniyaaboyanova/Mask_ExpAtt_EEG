```python
from ps_texture.texture_synthesis import *
import os 
```

```python
masking_paths = np.load("./Animacy/masking_paths.npy")
# Other params
num_depth = 5
num_ori = 4
num_neighbor = 7
iterations = 50 # what they have in orig paper
out_dir = "/projects/crunchie/boyanova/EEG_Things/Mask_ExpAtt_EEG/distractor_stimuli"
os.makedirs(out_dir, exist_ok=True)
```

```python
main_path = "/projects/crunchie/boyanova/EEG_Things/Mask_ExpAtt_EEG/target_stimuli"
categories = os.listdir(main_path)
masking_paths = []
for cat in categories:
    cat_path = os.path.join(main_path, cat)
    files = os.listdir(cat_path)
    masking_paths.extend([os.path.join(cat_path, x) for x in files])
masking_paths = np.array(masking_paths)
```

```python
# Image params
for i in range(len(masking_paths)):
    im_path = masking_paths[i]
    im = Image.open(im_path)
    im = im.resize((256, 256))
    im = np.array(im)
    im_name = im_path.split("/")[-1].split(".")[0]
    synthesis(im_name, im, im.shape[1], im.shape[0], num_depth, num_ori, num_neighbor, iterations, out_dir)
```
